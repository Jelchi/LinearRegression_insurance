```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# loading data
df = pd.read_csv('C:/Users/jelvi/OneDrive/Documents/PPTI BCA/KULIAH/CAWU 4/Machine Learning/TUT/Linear Regression/insurance.csv')
```

```python
# first look at data
print('shape: ', df.shape)
print(df)
```

```python
# visualizing data
sns.set_palette('magma')
sns.set_style ("darkgrid")
sns.regplot(x = 'age', y='charges', data=df)
plt.title('Age x Charges Distribution')
```

```python
Label_Encoder = LabelEncoder()
df['sex'] = Label_Encoder.fit_transform(df['sex'])
df['smoker'] = Label_Encoder.fit_transform(df['smoker'])
```

```python
df.drop(['region'], axis = 1, inplace=True)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
# preparing data for model
x = df['bmi'].values
x = x.reshape(-1, 1)
y= df['charges'].values
y = y.reshape(-1, 1)
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# building and applying linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
y_hat = model.predict(x_test)
y_hat
```

```python
# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_hat)
mse = mean_squared_error(y_test, y_hat)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_hat)
    
# Print the results
print("#" * 40)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
print("#" * 40)
```
